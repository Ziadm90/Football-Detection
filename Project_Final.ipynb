```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

cap = cv.VideoCapture("DataExample.mp4")

frame_count = 0

while frame_count<95:
    ret, frame = cap.read()
    
    if not ret:
        break

    frame_name = f"frame{frame_count}.jpg"
    
    frame_count += 1
    cv.imwrite(frame_name, frame)

cap.release()
```

Region label

```python
for i in range(95):

  img = cv.imread(f"frame{i}.jpg")

  lower_region = 50,130, 0
  upper_region = 255,255,128
  
  Region_mask = cv.inRange(img,lower_region,upper_region)

  plt.imshow(Region_mask)
  plt.show()
```

**yellow team label**

```python
for i in range(95):

  img = cv.imread(f"frame{i}.jpg")

  frame = cv.cvtColor(img, cv.COLOR_BGR2RGB)

  lower = 150,50,80

  upper = 255,255,255

  mask = cv.inRange(frame,lower,upper)

  plt.imshow(mask)
  plt.show()

```
